```python
from vega_datasets import data
import pandas as pd
import altair as alt
```

```python
cars = data.cars()
cars.head()
```

```python
alt.Chart(cars).mark_point().encode(
    alt.X('Miles_per_Gallon'),
    alt.Y('Horsepower')
)
```

```python
alt.theme.names() 
```

```python
alt.theme.enable('default')
```

```python
cars['URL'] = "https://google.com/search?q="+cars['Name']
cars.head()
```

```python
mensaje_tooltip = [alt.Tooltip('Horsepower',title="Caballos de fuerza:"),
                   alt.Tooltip('Miles_per_Gallon',title="Millas por galón:"),
                   alt.Tooltip('Name',title="Nombre:")]

autos = alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower'),
    alt.Y('Miles_per_Gallon'),
    href='URL',
    tooltip=mensaje_tooltip
).interactive()

autos
# autos.save('autos.html')
```

```python
#paso 1
brush = alt.selection_interval(resolve='intersect')

#paso 3
color_condition = alt.condition(brush,
                                'Origin',
                                alt.value('lightgray'))

autos = alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower'),
    alt.Y('Miles_per_Gallon'),
    color=color_condition,
    tooltip=mensaje_tooltip
).add_params(brush) #paso 2

autos
```

```python

autos_link = autos.encode(
    alt.X('Acceleration'),
    alt.Y('Weight_in_lbs')
)

autos | autos_link
```

```python
#1
click_legend = alt.selection_point(fields=['Origin'],bind='legend')
#3
# color_condition = alt.condition(click_legend,
#                                 'Origin',
#                                 alt.value('lightgray'))
opacity_condition = alt.condition(click_legend,
                                  alt.value(1),
                                  alt.value(0.2))


autos_leyenda = alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower'),
    alt.Y('Miles_per_Gallon'),
    # color=color_condition,
    color='Origin',
    opacity=opacity_condition,
    tooltip=mensaje_tooltip
).add_params(click_legend) #paso 2

autos_leyenda.properties(
    width=500,
    height=500
)
```

```python
#1
click = alt.selection_point(fields=['Origin'])
#3
bar_opacity_condition = alt.condition(click,
                                      alt.value(1),
                                      alt.value(0.2))

bars = alt.Chart(cars).mark_bar().encode(
    alt.X('count()'),
    alt.Y('Origin',sort='-x'),
    alt.Color('Origin'),
    opacity=bar_opacity_condition, #3
    tooltip=['count()']
).add_params(click) #2

puntos_opacity_condition = alt.condition(click,
                                         alt.value(1),
                                         alt.value(0.2))
puntos = alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower'),
    alt.Y('Miles_per_Gallon'),
    alt.Color('Origin'),
    opacity=puntos_opacity_condition
)

puntos & bars

```

```python



brush = alt.selection_interval() 

puntos_opacity_condition = alt.condition(brush,
                                         alt.value(1),
                                         alt.value(0.2))
puntos = alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower'),
    alt.Y('Miles_per_Gallon'),
    alt.Color('Origin'),
    opacity=puntos_opacity_condition
).add_params(brush)

bar_domain_max = cars.groupby('Origin').size().max()
# print(bar_domain_max)
bars = alt.Chart(cars).mark_bar().encode(
    alt.X('count()',axis=alt.Axis(grid=False),scale=alt.Scale(domain=[0,bar_domain_max])),
    alt.Y('Origin',sort='-x'),
    alt.Color('Origin'),
    tooltip=['count()']
).transform_filter(brush)

puntos & bars
```

```python
movies = pd.read_csv("https://raw.githubusercontent.com/erickedu85/dataset/master/movies-extended.csv", parse_dates=['Release_Date'])
movies
```

```python

alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    tooltip=['Title']
)

```

```python
unique_genres = sorted(movies['Major_Genre'].unique())
# print(unique_genres)
dropdown_genres = alt.binding_select(name='Seleccione el genero: ',options=unique_genres)

#1
selection_genre = alt.selection_point(fields=['Major_Genre'], bind=dropdown_genres)
#3
opacity_condition = alt.condition(selection_genre,
                                  alt.value(1),
                                  alt.value(0.2))

points = alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    opacity = opacity_condition,
    tooltip=['Title']
).add_params(selection_genre) #2

points
```

```python
unique_mpaa = sorted(movies['MPAA_Rating'].unique())
radiobuttons_mpaa = alt.binding_radio(name="Seleccione el MPAA Rating: ",options=unique_mpaa)


unique_genres = sorted(movies['Major_Genre'].unique())
# print(unique_genres)
dropdown_genres = alt.binding_select(name='Seleccione el genero: ',options=unique_genres)

#1
select_genre_and_mpaa = alt.selection_point(
    fields=['Major_Genre','MPAA_Rating'],
    bind={'Major_Genre':dropdown_genres,'MPAA_Rating':radiobuttons_mpaa}
) 

#3
opacity_condition = alt.condition(select_genre_and_mpaa,
                                  alt.value(1),
                                  alt.value(0.05))

points = alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    opacity = opacity_condition,
    tooltip=['Title']
).add_params(select_genre_and_mpaa) #2

points
```

```python
slider_imdb = alt.binding_range(name="Seleccione IMDB Rating", min=0, max=10, step=1)

imdb_selection = alt.selection_point(
    fields=['IMDB_Rating'],
    bind=slider_imdb
)
# opacity_condition = alt.condition(imdb_selection,
opacity_condition = alt.condition(alt.datum.IMDB_Rating <= imdb_selection.IMDB_Rating,
                                  alt.value(1),
                                  alt.value(0.1))

points = alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    opacity = opacity_condition,
    tooltip=['Title']
).add_params(imdb_selection).interactive()

points
```

```python
worldwide_gross_max = movies['Worldwide_Gross'].max()
# print(worldwide_gross_max)
slider_gross = alt.binding_range(
    name='Seleccione el Worldwide Gross: ', min=0, max=worldwide_gross_max, step=1e6
)

selection_worldwide_gross = alt.selection_point(
    fields=['Worldwide_Gross'],
    bind=slider_gross
)

opacity_condition = alt.condition(
    alt.datum.Worldwide_Gross >= selection_worldwide_gross.Worldwide_Gross,
    alt.value(1),
    alt.value(0.05)
)

points = alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    opacity = opacity_condition,
    tooltip=['Title','Worldwide_Gross']
).add_params(selection_worldwide_gross).interactive()

points
```

```python
brush = alt.selection_interval()

opacity_condition = alt.condition(brush,
                                  alt.value(1),
                                  alt.value(0.1))

bar_slider = alt.Chart(movies).mark_bar().encode(
  alt.X('Release_Date'),
  alt.Y('count()'),
  opacity = opacity_condition
).properties(
    width=500,
    height=50
).add_params(brush)

bar_slider


points = alt.Chart(movies).mark_circle().encode(
    alt.X('IMDB_Rating'),
    alt.Y('Rotten_Tomatoes_Rating'),
    alt.Color('Major_Genre'),
    opacity = opacity_condition,
    tooltip=['Title','Worldwide_Gross','Release_Date']
).properties(
    width=500,
    height=300
).interactive()
# .add_params(brush)

points & bar_slider
```
