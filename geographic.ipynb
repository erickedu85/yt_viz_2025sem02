```python
import pandas as pd
import geopandas as gpd
import altair as alt
from vega_datasets import data
```

```python
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"

gdf_ne = gpd.read_file(url)
gdf_ne = gdf_ne[['NAME','CONTINENT','POP_EST','geometry']]
gdf_ne
```

```python
alt.Chart(gdf_ne).mark_geoshape().properties(
    width=400,
    height=400
)
```

```python
gdf_sel = gdf_ne.query("CONTINENT == 'South America'")
print(gdf_sel)

alt.Chart(gdf_sel).mark_geoshape(
    stroke='lightgray',
    strokeWidth=0.5,
    fill='gray'
).properties(
    width=400,
    height=400
)
```

```python
alt.Chart(gdf_ne).mark_geoshape(
    stroke='lightgray',
    strokeWidth=0.5    
).transform_filter(
    alt.datum.CONTINENT == 'South America'
).properties(
    width=400,
    height=400
)
```

```python
alt.Chart(gdf_ne).mark_geoshape(
    stroke='lightgray',
    strokeWidth=0.5       
).project(
    scale=200,
    translate=[320,70]
).properties(
    width=400,
    height=400
)
```

```python
basemap = alt.Chart(gdf_sel).mark_geoshape(
    stroke='gray',
    strokeWidth=0.5,
    fill='lightgray'
).properties(
    width=400,
    height=400
)


bubbles = alt.Chart(gdf_sel).transform_calculate(
    centroid=alt.expr.geoCentroid(None,alt.datum)
).mark_circle(
    stroke='black'
).encode(
    longitude='centroid[0]:Q',
    latitude='centroid[1]:Q',
    size='POP_EST:Q',
    tooltip=['NAME:N','CONTINENT:N',alt.Tooltip('POP_EST:Q',format='.2f')]
)

basemap + bubbles
```

```python
alt.Chart(gdf_sel).mark_geoshape().encode(
    color='POP_EST:Q',
    tooltip=['NAME:N','CONTINENT:N',alt.Tooltip('POP_EST:Q',format='.2f')]
)
```

```python
#GEODATAFRAME CON GEOMETRIAS
gdf_us_counties = gpd.read_file(data.us_10m.url,driver='TopoJSON',layer='counties')
gdf_us_counties
```

```python
#DATAFRAME DE LA PROBLEMATICA
df_us_unemp = data.unemployment()
df_us_unemp
```

```python
alt.Chart(gdf_us_counties).mark_geoshape().transform_lookup(
    lookup='id',
    from_=alt.LookupData(data=df_us_unemp,key='id',fields=['rate'])
).encode(
    alt.Color('rate:Q'),
    tooltip=['id','rate:Q']
).project(
    type='albersUsa'
).properties(
    width=500,
    height=500
)
```
